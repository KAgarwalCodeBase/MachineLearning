## MatplotLib Tutorial
Matplotlib is a plotting libraray for the Python programming language and its numerical mathematics extension NumPy. It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like TkInter, WxPython, QT or GTK+.

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import numpy as np
```

```python
## Simple Examples
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
a = np.arange(40,50)
b = np.arange(50,60)
```

```python
## plotting using matplotlib
plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Graph in 2D')
plt.savefig('Test.png')
```

```python
y = x*x
```

```python
## plt plot 
plt.plot(x,y,'r*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D diagram')
```

```python
## Creating Subplots
plt.subplot(2,2,1)
plt.plot(x,y,'ro')
plt.subplot(2,2,2)
plt.plot(x,y,'b--')
plt.subplot(2,2,3)
plt.plot(x,y,'g*-')
plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

```python
## Complex graph using matplotlib
x = np.arange(1,11)
y = 3*x+5 # equation of straight line
plt.title('Matplotlib demo')
plt.xlabel('x axis caption')
plt.ylabel('y axis caption')
plt.plot(x,y)
plt.show()
```

```python
# compute the x and y coordinate for points on a sine curve
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form")

# plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
#subplot()
#Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

#Set up a subplot grid that has height 2 and width 1,
#and set the first such subplot as active.
plt.subplot(2,1,1)

#Make the first plot
plt.plot(x,y_sin,'r--')
plt.title('Sine')
plt.subplot(2,1,2)
plt.plot(x,y_cos,'g--')
plt.title('Cosine')
plt.show()
```

```python
## Bar plot
x = [2,8,10]
y = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('Bar graph')
plt.xlabel('X axis')
plt.ylabel('Y axis')
```

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
plt.hist(a)
plt.title("histogram")
```

## Box Plot using Matplotlib

```python
data  = [np.random.normal(0,std,100) for std in range(1,4)]

# rectangular box plot
plt.boxplot(data, vert=True, patch_artist=True)
```

```python
plt.boxplot(data, vert=False, patch_artist=False)
```

## Pie Chart

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0) # explode 1st slice

# plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
       autopct='%1.1f%%', shadow=True)
plt.axis('equal')
plt.show()
```
