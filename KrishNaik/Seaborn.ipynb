### Correlation with HeatMap
A correction heatmap uses colored cells, typically in a monchromatic scale, to show a 2D correction matrix (table) between twp discrete dimensions or event types. It is very important in Feature Selection.

```python
# Selecting only numeric data, i.e excluding the categorical data.
numeric_df = df.select_dtypes(include='number')
numeric_df.corr()
```

```python
# Creating the heatmap which helps to analyze the correlation between different attributes.
sns.heatmap(numeric_df.corr())
```

## JoinPlot
A join plot allows to study the relationship between 2 numeric variables. The central chart display their correlation. It is usually a scatterplot, a hexbin plot, a 2D histogram or a 2D density plot

### Univariate Analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip', y='total_bill',data=df,kind='reg')
```

### Pair Plot

A 'pairs plot' is also known as scatterplot, in which one variable in the same data row is matched with another variable's value, like this: Pairs plots are just eloborations on this, showing all variables paired with all the other variables.


```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue='smoker')
```

## Dist plot
Dist plot helps us to check the distribution of the columns feature

```python
sns.histplot(df['tip'], label='Tip', bins=10)
```

```python
sns.distplot(df['tip'], kde=False, bins=10)
```

```python
sns.distplot(df['total_bill'], label='Total Bill')
```

```python
import matplotlib.pyplot as plt
sns.histplot(df['total_bill'], label='Total Bill')
sns.histplot(df['tip'],label='Tip')
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Comparison between Tip and Total Bill Distribution')
plt.legend()
plt.show()
```

## Categorical Plot
Seaborn also helps us in doing the analysis on Categorical Data points.In this section we will discuss about
- boxplot
- violinplot
- countplot
- bar plot


```python
## Count plot
sns.countplot(x='sex',data=df)
```

```python
## Count plot 
sns.countplot(y='sex', data=df)
```

```python
## Bar plot
sns.barplot(x='total_bill', y='sex', data=df)
```

```python
## Bar plot
sns.barplot(x= 'smoker',y='total_bill',data=df)
```

## Box plot
A box and whisker plot (sometimes called a boxplot) is a graph that presents information from a five-number summary.

```python
sns.boxplot(x = 'smoker', y = 'total_bill', data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='h')
```

```python
# Categorize my data based on some other categories
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

## Violin plot
Violin plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot

```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
